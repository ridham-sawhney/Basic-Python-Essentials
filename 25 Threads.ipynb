Threads

```python
import threading
import time
```

```python
def longSquare(num, results):
    time.sleep(1)
    results[num] = num**2

results = {}
t1 = threading.Thread(target=longSquare, args=(2,results))
t2 = threading.Thread(target=longSquare, args=(4,results))

t1.start()
t2.start()

t1.join()
t2.join()

print(results)
```

```python
def longSquare(num, results):
    time.sleep(1)
    results[num] = num**2

results = {}
threads = [threading.Thread(target=longSquare, args=(n,results)) for n in range(1,100)]

[t.start() for t in threads]
[t.join() for t in threads]


print(results)
```
