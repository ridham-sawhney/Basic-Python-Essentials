```python
def allPrimesUpTo(n):
    foundPrimes = [];
    
    for num in range(2,n):
        # print(f'checking for {num}')
        for foundPrime in foundPrimes:
            if (foundPrime > int(num**0.5)+1):
                continue
                
            if (num % foundPrime == 0):
                break
        else:
            # print(f"Didnot able to find from found primes:{foundPrimes}")
            for i in range(2,int(num**0.5)+1):
                if num % i == 0:
                    break
            else:
                foundPrimes.append(num)
        
    return foundPrimes
```

```python
def allPrimesUpTo_optimized(n):
    foundPrimes = [2];
    
    for num in range(3,n):
        sqrt = num**0.5
        for foundPrime in foundPrimes:
            if (num % foundPrime == 0):
                break #not a prime
            if (foundPrime > sqrt):
                foundPrimes.append(num)
                break
    return foundPrimes
```

```python
allPrimesUpTo_optimized(100)
```
