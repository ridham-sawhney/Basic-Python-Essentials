Functions

```python
def perform_operation(a, b, operation):
    if operation == "add":
        return a + b
    elif operation == "subtract":
        return a - b
    elif operation == "multiply": a * b
    elif operation == "divide":
        if b == 0:
            raise ValueError("Cannot divide by zero")
        return a / b
    else:
        raise ValueError("Invalid operation")
    
```

```python
perform_operation(1,2,"add")
```

Named parameters


```python
def perform_operation(a, b, operation='add'):
    if operation == "add":
        return a + b
    elif operation == "subtract":
        return a - b
    elif operation == "multiply": a * b
    elif operation == "divide":
        if b == 0:
            raise ValueError("Cannot divide by zero")
        return a / b
    else:
        raise ValueError("Invalid operation")
    
```

```python
perform_operation(1,2)
```

args


```python
def perform_operation(*args):
    print(args)
    
perform_operation(1,3,4,5)    
perform_operation(1,3,4,5,5,6)    
```

```python
def perform_operation(*args,**kwargs):
    print(args)
    print(kwargs)
    
perform_operation(1,3,4,5,operation="sum")    
perform_operation(1,3,4,5,5,6)  
```

```python
import math
```

```python
def perform_operation(*args,**kwargs):
    operation = kwargs.get('operation')
    if operation=='sum':
        return sum(args)
    elif operation=='product':
        return math.prod(args)
    else:
        return "Invalid Operation"
    
print(perform_operation(1,3,4,5,operation='sum'))
print(perform_operation(1,3,4,operation='product')) 
```

locals()

```python
def perform_operation(*args,**kwargs):
    print(locals())
    
perform_operation(1,3,4,5,operation="sum")    
perform_operation(1,3,4,5,5,6)  
```

```python
def perform_operation(val1,val2):
    print(locals())
    
perform_operation(1,3)  
```

globals()

```python
def perform_operation(*args,**kwargs):
    print(globals())
    
perform_operation(1,3,4,5,operation="sum")    
perform_operation(1,3,4,5,5,6)  
```

```python
def toLowerCase(s):
    return s.lower()

def removePunctuation(s):
    return s.replace(".","").replace(",","")
```

```python
processingFunctions = [toLowerCase,removePunctuation]
text = '''
Hi there! How are you doing today? The weather is great, isn't it?
I'm doing well, thank you.

'''
for func in processingFunctions:
   text = func(text)
   
print(text) 
```
