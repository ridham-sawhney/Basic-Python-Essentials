AAAAABB <-> [('A',5),('B',2)]

```python
def encodeString(stringVal):
    encodedValues = []
    prevChar=stringVal[0]
    count = 0
    for char in stringVal:
        if(prevChar!=char):
            encodedValues.append((prevChar,count))
            prevChar = char
            count = 0
        count=count+1

    encodedValues.append((prevChar,count))
    return encodedValues

def decodeString(encodedList):
    finalString = ''
    for char,count in encodedList:
        finalString = finalString+ char*count
    return finalString
```

```python
encodedString = encodeString('AAAAABB')
print(encodedString)
decodedString = decodeString(encodedString)
print(decodedString)
```
