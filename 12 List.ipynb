Lists


```python
myList = [1,2,3,4]
```

```python
print(myList[0:2])
print(myList[0:2:3])
print(myList[::2]) # 2 is a step here
print(myList[::-1])
```

```python
myList = list(range(100))
print(myList[::10])
```

Modifying List

```python
a = [1,2,3,4]
b=a
a.append(5)
print(a)
print(b)
```

```python
a = [1,2,3,4]
b=a.copy()
a.append(5)
print(a)
print(b)
```

```python
list1 = [1,2,3,4,5]
print(list1.pop())
print(list1)
```

```python
list1.remove(12) #if item doenot exist it throws an error
```

```python
list1.count(10) # gives count of element in the list , if there is not returns 0
```
