```python

with open("./Files/Alaska.csv",'w',newline='') as f:
    writer = csv.DictWriter(f,fieldnames=data[0].keys(),delimiter='|')
    writer.writeheader()
    for row in [row for row in data if row['state'] == 'Alaska']:
        writer.writerow(row)
```

JSON

```python
import json
from json import JSONEncoder
```

```python
jsonString = '{"b":"bear","c":"cat","d":"dog"}'
a= json.loads(jsonString)
print(a.get("b"))
```

```python
jsonObj = {"b":"bear","c":"cat","d":"dog"}
print(json.dumps(jsonObj))
```

```python
class Animal:
    def __init__(self,name):
        self.name = name
        
class AnimalEncoder(JSONEncoder):
    def default(self,o):
        if type(o) is Animal:
            return o.name
        return super().default(o)
    
jsonObj = {"b":Animal("bear"),"c":Animal("cat")}
json.dumps(jsonObj,cls=AnimalEncoder)
```
