Process

```python
from multiprocessing import Process
import time
```

```python
def longSquare(num, results):
    time.sleep(1)
    print(num**2)
    print("Finished Processing")

results = {}
p1 = Process(target=longSquare, args=(2,results))
p2 = Process(target=longSquare, args=(4,results))

p1.start()
p2.start()

p1.join()
p2.join()
```
